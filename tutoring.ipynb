```python
df['N%'] = 0

for i in range(len(df['N%'])):  
    a = df['%(จำนวนนักศึกษาที่จดทะเบียน)'][i].split('.')[0]
    b = df['%(จำนวนนักศึกษาที่จดทะเบียน)'][i].split('.')[1]
    a = a.replace('(' , '')
    a += '.'
    b = b.replace('%)' , '')
    df.iloc[i,29] = float(''.join([a,b])) # ใส่เลขcolumn N% ตรง 29
```
